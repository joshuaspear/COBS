```python
import pandas as pd
import pickle
from functools import reduce
from ope.config import TRACKER_PATH, EXPERIMENT_DIR
from pymlrf.ModelTracking import SerialisedTracker
from typing import Dict, List
```

```python
tracker = SerialisedTracker(path=TRACKER_PATH)
tracker.read()
```

```python
path = "./toy_graph.pkl"
with open(path, "rb") as f:
    results = pickle.load(f)
```

```python
def summarise_seed_results(results:List[Dict])->pd.DataFrame:
    res_df = []
    for seed in range(len(results)):
        _tmp = pd.DataFrame(
            list(results[seed].values())[:-1],
            columns=[f"val_{seed}",f"diff_to_on_policy_{seed}"],
            index=list(results[seed].keys())[:-1]
            )
        _tmp[f"diff_to_on_policy_rank_{seed}"] = _tmp[f"diff_to_on_policy_{seed}"].rank()
        res_df.append(_tmp)
    res_df = reduce(lambda x,y: pd.merge(x,y,how="inner",right_index=True, left_index=True),res_df)
    return res_df
    
```

```python
rank_cols = [col for col in res_df.columns if "rank" in col]
res_df["mean_rank"] = res_df[rank_cols].mean(axis=1)
res_df["std_rank"] = res_df[rank_cols].std(axis=1)
```
